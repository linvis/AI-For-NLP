* precision， recall，f1, auc 分别是什么意思？ 假设一个城市有 10000 人，有 30 个犯罪分子，警察抓到了 35 个人，其中 20 个是犯罪分子，请问这个警察的 precision, recall, f1,auc 分别是什么？

    precision - 预测对的占预测总数的比例
    
    recall - 预测对的，占实际对的比例

    f1 = $\frac{2*P*R}{P + R}$

    precision = 20 / 35
    recall = 20 / 30
    f1 = 0.7619 / 1.238
    就一次测试auc没法算吧.
