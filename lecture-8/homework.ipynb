```python
def clean_content(content):
    content = re.findall(re.compile('[\w|\d]+'), str(content))
    return ' '.join(content)
```

```python
def prepare_data(contents, num):
    
    content = contents['content'].tolist()
    source = contents['source'].tolist()
    
    random_content = []
    random_source = []
    
    for i in range(num):
        random_idx = random.randint(1, num)
        random_content += [clean_content(content[random_idx])]
        random_source += [source[random_idx]]
    
    return random_content, random_source
```

```python
from collections import Counter
```

```python
def to_chunks(_list):
    chunks = []
    for v in _list:
        for t in v:
            chunks.append(t)
    return chunks
```

```python
def prepare_word_freq(contents):
    content = contents['content'].tolist()
    content = [clean_content(con) for con in content]
    
    content = [list(' '.join(jieba.cut(con)).split()) for con in content]
    
    content = to_chunks(content)
    counter = Counter(content)
    
    return counter
```

```python
class Sen2vec:
    def __init__(self, model, word_freq):
        self.scalar_a = 0.0001

        # load word frequency data
        self.word_freq = word_freq

        self.word_vec = model
        
        self.word_sum = 0
        for k, v in word_freq.items():
            self.word_sum += int(v)

    def get_vec(self, sentences):
        # cut to small sentence
        sentences = clean_content(sentences)
        tokens = ' '.join(jieba.cut(sentences)).split()

        sen_vec = 0
        for token in tokens:
            freq = self.word_freq[token]
            vec = self.word_vec.wv[token]
            # print("word:{} freq: {}".format(token, freq))
            # print("word vec: {}, type {}".format(vec, type(vec)))

            sen_vec += self.scalar_a / (self.scalar_a + freq / self.word_sum) * vec
        
        return sen_vec

    def get_cos(self, v1, v2):
        return cosine(v1, v2)
```

```python
class SetType:
    def __init__(self):
        self.text = None
        self.source = None
        self.index = None
        self.vector = None
        self.y = None
```

```python
class DataSet:
    def __init__(self, corpus, words_model, words_counter,\
                 train_scale=0.6, test_scale=0.2, valid_scale=0.2):
        self.corpus = corpus
        if (train_scale + test_scale + valid_scale > 1):
            print("Error: sum of train, test, valid is bigger than 1")
            return 
        self.content = None
        self.source = None
        self.content_xinhuashe = []
        self.content_not_xinhuashe = []
        
        self.train_scale = train_scale
        self.test_scale = test_scale
        self.valid_scale = valid_scale
        
        self.train_text = None
        self.train = None
        self.test_text = None
        self.test = None
        self.valid_text = None
        self.valid = None
        self.sen2vec = Sen2vec(words_model, words_counter)
        
        self.prepare_data(self.corpus)
        
        
        
    def clean_content(self, content):
        content = re.findall(re.compile('[\w|\d]+'), str(content))
        return ' '.join(content)
    
    def prepare_data(self, corpus):
    
        content = corpus['content'].tolist()
        self.content = [clean_content(con) for con in content]
        self.source = corpus['source'].tolist()
        
        for i in range(len(self.source)):
            data = SetType()
            data.text = self.content[i]
            data.index = i
            data.source = self.source[i]
            data.vector = self.sen2vec.get_vec(data.text)
            if '新华社' in self.source[i]:
                data.y = 1
                self.content_xinhuashe.append(data)
            else:
                data.y = 0
                self.content_xinhuashe.append(data)
                
        #print(self.content_xinhuashe)
        #print(self.content_not_xinhuashe)
                
        
    def get_random_sample(self, sample, scale):
        print(range(len(sample)))
        print(int(len(sample) * scale))
        print(scale)
        rand = random.sample(range(len(sample)), int(len(sample) * scale))
        
        rand_sample = []
        remain_sample = []
        for num in rand:
            rand_sample.append(sample[num])
        
        remain_num = list(set(list(i for i in range(len(sample)))) - set(rand))
        for num in remain_num:
            remain_sample.append(sample[num])
        
        return rand_sample, remain_sample
    
    def get_set(self):
        
        # get train
        print(int(len(self.content_xinhuashe) * self.train_scale))
        train_xinhuashe, remain_xinhuashe = self.get_random_sample(self.content_xinhuashe, self.train_scale)
        
        train_not_xinhuashe, remain_not_xinhuashe = self.get_random_sample(self.content_not_xinhuashe, self.train_scale)
        
        self.train = train_xinhuashe + train_not_xinhuashe
        
        # get text
        test_scale = self.test_scale / (self.test_scale + self.valid_scale)
        test_xinhuashe, remain_xinhuashe = self.get_random_sample(remain_xinhuashe, test_scale)
        
        
        test_not_xinhuashe, remain_not_xinhuashe = self.get_random_sample(remain_not_xinhuashe, test_scale)
        self.test_text = [[1, tra] for tra in test_xinhuashe] \
            + [[0, tra] for tra in test_not_xinhuashe]
        
        self.test = test_xinhuashe + test_not_xinhuashe
        
        # get valid 
        self.valid = remain_xinhuashe + remain_not_xinhuashe

        return self.train, self.test, self.valid
    
    def get_train_text(self):
        return self.train_text
    
    def get_test_text(self):
        return self.test_text
    
    def get_valid_text(self):
        return self.valid_text
        
```
